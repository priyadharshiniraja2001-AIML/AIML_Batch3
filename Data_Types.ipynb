```python
# with open "a--> append","r--> read","w-->write"

with open ('filename100.pdf',mode = 'w') as f:
    f.write("hello world !!")
    f.write("\n")
    f.write("Welcome to our class DS,\n")
```

```python
with open ("filename100.pdf",'a') as f:
    f.write("hello how are you ,'I am good',\n")
```

```python
with open("filename100.pdf",mode = 'r') as f:
    data = f.read()
print(data)
```

# Loops

```python
data = [15,10,20,14,30,25]
new_data = []
for i in data:
    result = i+2
    new_data.append(result)
print(new_data)
```

```python
for i in range(5,15):
    print(i)
```
